```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11) #11 значень від 0 до 5
y=x**2
```

```python
x
```

```python
y
```

```python
#functional method
plt.plot(x,y) 
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('title')
```

```python
#create multiply plots
plt.subplot(1, 2, 1) #1 рядок, 2 стовпчики, 1й графік
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2) #1 рядок, 2 стовпчики, 2й графік
plt.plot(x, y, 'g')
```

```python
#oo method
fig=plt.figure()

axes=fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('tittle')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0, 0, 1, 1])
axes2=fig.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('tittle')

axes2.plot(y, x, 'g')
axes2.set_xlabel('Y label')
axes2.set_ylabel('X label')
axes2.set_title('tittle')
```

```python
#ініціалізація графіка як об'єкта
fig, axes=plt.subplots()

axes.plot(y, x, 'b')
axes.set_xlabel('Y label')
axes.set_ylabel('X label')
axes.set_title('tittle')
```

```python
fig, axes=plt.subplots(nrows=2, ncols=3)

axes[0,0].plot(x,y,'g')
axes[0,1].plot(y,x, 'b')
axes[0,2].plot(x,y, 'y')

axes[1,0].plot(x,y,'g')
axes[1,1].plot(y,x, 'b')
axes[1,2].plot(x,y,'y')
```

```python
fig, axes=plt.subplots(nrows=4, ncols=4)
```

```python
plt.tight_layout()
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

```python
fig=plt.figure(figsize=(8,4), dpi=100)
```

```python
#matlab style line color & style
fig, ax=plt.subplots()
ax.plot(x, x**2, color='y')
ax.plot(x, x**3, color='g')
```

```python
#matlab style ширина линии
fig, ax=plt.subplots()
ax.plot(x, x**2, lw=2, ls='--' )
```

```python
#matlab style тип, размер и цвет маркеров
fig, ax=plt.subplots()
ax.plot(x, x**2, marker='^', markersize=10, markerfacecolor='b')
```
