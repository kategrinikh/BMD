```python
import seaborn as sns
%matplotlib inline
```

```python
dots=sns.load_dataset('dots')
```

```python
dots.head()
```

```python
sns.distplot(dots['time'])
```

```python
sns.jointplot(x='time', y='firing_rate', data=dots)
```

```python
sns.pairplot(dots)
```

```python
sns.pairplot(dots, hue='choice')
```

```python
sns.barplot(x='choice', y='coherence', data=dots)
```

```python
import numpy as np
sns.barplot(x='choice', y='coherence', data=dots, estimator=np.median)
```

```python
sns.boxplot(x='coherence', y='time', data=dots, palette='rainbow')
```

```python
sns.stripplot(x='coherence', y='time', data=dots,)
```

```python
sns.heatmap(dots.corr())
```

```python
#grid
g=sns.FacetGrid(dots, col="align", row="choice")
```
